```python
### BAR CHART & LINE PLOT Of Percent of Bachelor Degrees in Computer Science Women Hold ###

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bachelor_data = pd.read_csv(r"C:\Users\rames\datasets_6115_9007_percent-bachelors-degrees-women-usa.csv", delimiter=',')

# Data to plot in Bar and Line
plt.bar(x=bachelor_data['Year'],height=bachelor_data['Computer Science'], color="pink")
plt.plot(bachelor_data['Year'], bachelor_data['Computer Science'], color="black")

# Plot
plt.title("Women in CS in the U.S.")
plt.xlabel("Year")
plt.ylabel("Percent of U.S. Bachelor CS Degrees Women Hold")
plt.show()
```

```python
### LINE PLOTS OF Average Percent of Bachelor Degrees in STEM Women Hold compared to Individual STEM Degrees ###

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bachelor_data = pd.read_csv(r"C:\Users\rames\datasets_6115_9007_percent-bachelors-degrees-women-usa.csv", delimiter=',')

# STEM = Architecture , Biology, Computer Science, Engineering, Math and Statistics, Physical Sciences 
num_stem = []
for i in range(42):
    avg = (bachelor_data['Architecture'][i] + bachelor_data['Biology'][i] + 
           bachelor_data['Computer Science'][i] + bachelor_data['Engineering'][i] + 
           bachelor_data['Math and Statistics'][i] + bachelor_data['Physical Sciences'][i]) / 6
    num_stem.append(avg)

#Line plot
plt.plot(bachelor_data['Year'], num_stem, color="red", label="STEM Average")
plt.plot(bachelor_data['Year'], bachelor_data['Architecture'], color="orange", label="Architecture")
plt.plot(bachelor_data['Year'], bachelor_data['Biology'], color="yellow", label="Biology")
plt.plot(bachelor_data['Year'], bachelor_data['Computer Science'], color="green", label="Computer Science")
plt.plot(bachelor_data['Year'], bachelor_data['Engineering'], color="blue", label="Engineering")
plt.plot(bachelor_data['Year'], bachelor_data['Math and Statistics'], color="indigo", label="Math and Statistics")
plt.plot(bachelor_data['Year'], bachelor_data['Physical Sciences'], color="violet", label="Physical Sciences")

# Put a legend to the right of the current axis
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Plot
plt.title("Women in STEM in the U.S.")
plt.xlabel("Year")
plt.ylabel("Percent of U.S. Bachelor STEM Degrees Women Hold")
plt.show()
```
